```python
vowels="aieou"
Name="Moamen Eltony"
Name=Name.lower()
c=0
for i in Name:
  if i in vowels:
    c+=1
print(c)
```

```python
sentence="We love iti iti iti"
sentence=sentence.lower()
c=0
for i in range(0,len(sentence)-2):
  if 'iti'==sentence[i:i+3]:
    c+=1
print(c)
```

```python
vowels="aieou"
Name="Moamen Eltony"
Name=Name.lower()
NewName=""
for i in Name:
  if i in vowels:
    continue
  else:
    NewName+=i
print(NewName)
```

```python
num=input("Enter Number")
res="["
if num.isnumeric():
  num=int(num)
  for i in range(1,num+1):
    res=res+'['
    for j in range(1,i+1):
      if j!=i:
        res=res+(f"{i*j},")
      else:
        res=res+(f"{i*j}]")
  res=res+']'
  print(res)
else:
  print("Invalid Input")
```

```python
num=input("Enter Number")
if num.isnumeric():
  num=int(num)
  for i in range(1,num):
    print((" "*(num-1-i))+("*"*i))
else:
  print("Invalid Input")
```
