```python
def ListGenerator(start,length):
  if type(start)==int and type(length)==int:
    return list(range(start,start+length))
  else:
    print("Invalid Arguments")


print(ListGenerator(1,5))
```

```python
def FizzBuzz(num):
  if type(num)==int:
    res=""
    if num%3==0:
      res+='Fizz'
    if num%5==0:
      res+='Buzz'
    return res
  else:
    print("Invalid Argument")

print(FizzBuzz(15))
```

```python
def StrReverse(ActualStr):
  if type(ActualStr)==str:
    return ActualStr[::-1]
  else:
    print("Invalid Argument")

print(StrReverse("Moamen Eltony"))
```

```python
def InformationGetter():
  Name=input("Enter Your Name:")
  if Name and Name.replace(' ',"").isalpha():
    Email=input("Enter Your Email:")
    flag=False
    for i in Email:
      if i=='@':
        flag=True
      if flag and i=='.':
        break
    else:
      print("Invalid Email")
      return

    print(f'Your Name Is:{Name}')
    print(f'Your Email Is:{Email}')

  else:
    print("Invalid Name")


InformationGetter()
```

```python
def LongestSubstring(ActualStr):
  if type(ActualStr)==str and ActualStr.replace(' ',"").isalpha():
    ActualStr=ActualStr.replace(' ','')
    ActualStr=ActualStr.lower()
    res=""
    tres=""+ActualStr[0]
    for i in range(1,len(ActualStr)):
      if ActualStr[i-1]<ActualStr[i]:
        tres+=ActualStr[i]
      else:
        if len(tres)>len(res):
          res=tres
        tres=""+ActualStr[i]

    if len(tres)>len(res):
      res=tres
    return res

  else:
    print("Invalid Argument")

print(LongestSubstring('abdulrahmanbqruv' ))

```

```python
def sum():
  num=''
  sum=0
  while True:
    num=input("Enter A Number:")
    if num and num.isalpha() and num.lower()=='done':
      break
    if num.isdigit():
      sum+=int(num)
    else:
      print("Invalid Input")

  return sum

print(sum())
```

```python
import warnings
import random

warnings.filterwarnings("ignore")


def WordGuessingGame():
    HANGMANPICS = [r'''
  +---+
  |   |
      |
      |
      |
      |
=========''', r'''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', r'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', r'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', r'''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', r'''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', r'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']

    Words = ["tree", "home", "city", "field", "bannana", "padel", "ai"]

    Name = input("Enter Your Name: ")
    if Name and Name.replace(' ', "").isalpha():
        print(f'Welcome {Name}')

        ChosenWord = list(Words[random.randint(0, 6)])
        CorrectChars = ['_' for _ in ChosenWord]

        Counter = 0
        WrongcharsCounter = 0

        for i in range(7):
            print(" ".join(CorrectChars))
            Guess = input("Enter Your Guess: ")

            if Guess and Guess.isalpha() and len(Guess) == 1 and Guess.lower() in ChosenWord:
                print("Correct")

                for j in range(len(ChosenWord)):
                    if ChosenWord[j] == Guess.lower() and CorrectChars[j] == '_':
                        CorrectChars[j] = Guess.lower()
                        ChosenWord[j] = '_'
                        Counter += 1

                if Counter == len(CorrectChars):
                    print("You Won")
                    break
            else:
                print("Wrong Guess")
                WrongcharsCounter += 1
                print(HANGMANPICS[WrongcharsCounter])
        else:
            print("You Lose")
            if WrongcharsCounter != 6:
                print(HANGMANPICS[6])
    else:
        print("Invalid Name")


WordGuessingGame()
```
